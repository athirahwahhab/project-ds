```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('survey lung cancer.csv')
data
```

```python
data.info
```

```python
mean = data['AGE'].mean()
median= data['AGE'].median()
mode = data['AGE'].mode()
std_dev= data['AGE'].std()
print("Mean of age ")
print(round(mean,5))
print("Median of age")
print(round(median,5))
print("Mode of age")
print(round(mode,5))
print("Standard deviation of Age")
print(round(std_dev,5))
```

```python
mean = data['SMOKING'].mean()
median= data['SMOKING'].median()
mode = data['SMOKING'].mode()
std_dev= data['SMOKING'].std()
print("Mean of smoking ")
print(round(mean,5))
print("Median of smoking")
print(round(median,5))
print("Mode of smoking")
print(round(mode,5))
print("Standard deviation of smoking")
print(round(std_dev,5))
```

```python
mean = data['SWALLOWING DIFFICULTY'].mean()
median= data['SWALLOWING DIFFICULTY'].median()
mode = data['SWALLOWING DIFFICULTY'].mode()
std_dev= data['SWALLOWING DIFFICULTY'].std()
print("Mean of Swallowing difficulty ")
print(round(mean,5))
print("Median of Swallowing difficulty")
print(round(median,5))
print("Mode of Swallowing difficulty")
print(round(mode,5))
print("Standard deviation of Swallowing difficulty")
print(round(std_dev,5))
```
